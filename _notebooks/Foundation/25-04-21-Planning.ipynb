# Biotech Innovation Engagement Project

## Ideation
The Scripps Research Biotech Innovation Engagement project aims to increase public awareness of Scripps Research's biotech advancements through an online game. By integrating machine learning and data analysis, we plan to create an experience that educates social media followers on biotech concepts. The project aligns with Scripps Research’s mission to promote scientific literacy and encourage interest in biotechnology among the public.  

## Why it interests us
This project is exciting because it blends interactive technology, biotech education, and social media engagement. Many people are unaware of the latest innovations in biotech, and through this game, we can make complex topics accessible and engaging. Our goal is to increase engagement by 30% within three months, ensuring that more people interact with and understand the impact of biotech research. The initiative also demonstrates how gamification and data-driven insights can enhance science communication.  

## Data Sets
The project will utilize biotech research data from Scripps Research publications, public databases, and other reports. Team members will analyze these datasets to extract Scripps's biotech breakthroughs. Machine learning tinkers, such as adaptive and repetitive question generation, will be explored to improve engagement and knowledge retention.  

Example Dataset: https://catalog.data.gov/dataset/biotechnology-regulatory-services-public-data-file-permits-notifications-and-petitions

This dataset provides data on Biotechnology Regulatory Services. This shows which hospitals have licenses and services to do Biotechnological research and care. Along with the type of license.

# Project Name: Bio-SKANZ

## Objective
Develop an interactive online simulation that leverages machine learning to educate Scripps Research’s social media audience about biotech innovations. The goal is to increase engagement by 30% through:  

- A virtual lab simulation allowing users to experiment with biotech techniques.  
- A biotech trivia challenge with adaptive difficulty based on user performance.  
- An innovation discovery quest guiding users through real-world biotech breakthroughs.  

This project aims to showcase Scripps Research’s contributions to biotech while fostering an interactive and educational experience.

# Project Phases

### Week 1-2: Research & Data Collection
   - Gather biotech innovation data from Scripps Research and public datasets.  
   - Analyze data sources for key scientific concepts.  
   - Identify features for machine learning models.  

### Week 3-5: Game Development & AI Integration
   - Implement adaptive question generation & machine learning analytics.  
   - Develop interactive biotech simulations and quest mechanics.  

### Week 6-7: Integration & Testing
   - Combine AI-driven features with game design.  
   - Test user engagement levels and refine experience.  

### Week 8-9: Deployment & Evaluation (Week 8-9)  
   - Deploy game on social media platforms.  
   - Measure user engagement against the 30% increase target.  
   - Adjust game mechanics based on analytics feedback.  

### Week 10: Final Presentation  
   - Showcase the interactive game to Pilot City Scripps Research.  
   - Document feedback and propose future improvements.
